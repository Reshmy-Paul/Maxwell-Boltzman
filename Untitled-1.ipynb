```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import k, m_u # k is the boltzmann constant , m_u is the used for conversion from molar mass to molecular mass

# Constants
temperature_160K = 160  # in Kelvin
temperature_300K = 300  # in Kelvin
temperature_2000k = 2000 # Temp in Kelvin
mass_CO2 = 44 * m_u  # CO2 molecular mass (amu)
boltzmann_constant = k

# Maxwell-Boltzmann distribution formula
def maxwell_boltzmann_distribution(v, T, m):
    coefficient = np.sqrt((m / (2 * np.pi * boltzmann_constant * T))**3)
    exponential = np.exp(-m * v**2 / (2 * boltzmann_constant * T))
    return 4 * np.pi * v**2 * coefficient * exponential

# Velocity range for X axis
v = np.linspace(0, 3000, 1000)  # in m/s

# Distributions for each temperatures
distribution_160K = maxwell_boltzmann_distribution(v, temperature_160K, mass_CO2)
distribution_300K = maxwell_boltzmann_distribution(v, temperature_300K, mass_CO2)
distribution_2000k = maxwell_boltzmann_distribution(v,temperature_2000k,mass_CO2)

# Plot for the distributions for different temperature range
plt.figure(figsize=(12, 6))
plt.plot(v, distribution_160K, label='CO2 at 160 K', color='blue')
plt.plot(v, distribution_300K, label='CO2 at 300 K', color='red')
plt.plot(v,distribution_2000k,label = 'CO2 at 2000 K', color='green')
plt.xlabel('Velocity (m/s)')
plt.ylabel('Probability Density')
plt.title('Maxwell-Boltzmann Distribution for CO2 at 160K, 300K and 2000k')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Molecular mass calculations for each compound
mass_3N2 = 84.06 * m_u #3N2 molecular mass(amu)
mass_He = 4.0026 * m_u #He molecular mass(amu)
mass_H = 1.008 * m_u #H molecular mass(amu)

# Distributions for each compound
distribution_3N2 = maxwell_boltzmann_distribution(v, temperature_300K, mass_3N2)
distribution_He = maxwell_boltzmann_distribution(v, temperature_300K, mass_He)
distribution_H = maxwell_boltzmann_distribution(v, temperature_300K, mass_H)

# Velocity range for X axis
v = np.linspace(0, 10000, 1000)  # in m/s

# Plot for the distributions for different compounds at 300K
plt.figure(figsize=(12,6))
plt.plot(v, distribution_3N2, label='3N2 at 300 K', color='blue')
plt.plot(v, distribution_He, label='He at 300 K', color='red')
plt.plot(v,distribution_H, label = 'H at 300 K', color='green')
plt.xlabel('Velocity (m/s)')
plt.ylabel('Probability Density')
plt.title('Maxwell-Boltzmann Distribution for 3N2, He and H at 300K')
plt.legend()
plt.grid(True)
plt.show()
```
